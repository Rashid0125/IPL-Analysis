# Finding Season in which Match have been Played

```python
match_data['Season'] = pd.DatetimeIndex(match_data['date']).year
match_data.head()
```

# Find the Number of matches Played in every Season

```python
match_per_season = match_data.groupby('Season')['id'].count().reset_index().rename(columns = {'id':'Matches'})
match_per_season
```

```python
sns.barplot(data=match_per_season,x='Season',y='Matches')
```

```python
sns.countplot(match_data['Season'])
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('IPL Season',fontsize=14)
plt.ylabel('Total no. of Matches',fontsize=14)
plt.title('Total Matches played in each Season',fontsize=14,fontweight="bold")
plt.show()
```

# Season-wise Data

```python
season_data = match_data[['id','Season']].merge(ball_data,left_on='id',right_on='id',how='left').drop('id',axis=1)
```

```python
season_data
```

# Total Run Score in each Season

```python
season = season_data.groupby(['Season'])['total_runs'].sum().reset_index()
season
```

```python
p = season.set_index('Season')
ax = plt.axes()
ax.set(facecolor = 'violet')

sns.lineplot(data=p,palette = 'magma')
plt.title('Total runs in each season',fontsize=14,fontweight="bold")
plt.show()
```

# Obserevation
Maximum Number of runs is scored in Season 2012

# Average Run Scored in Each Match

```python
runs_per_season=pd.concat([match_per_season,season.iloc[:,1]],axis=1)

runs_per_season
```

```python
runs_per_season['Run scored per Match'] = runs_per_season['total_runs']/runs_per_season['Matches']
#runs_per_season.set_index('Season',inplace=True)
runs_per_season
```
